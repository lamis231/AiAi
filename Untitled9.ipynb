```python
import pandas as pd, numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score, average_precision_score

# 1) اللابل
positive = {"book","favorite","like"}
df["label"] = df["action"].isin(positive).astype(int)

# 2) دوال مساعدة
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1; dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 2*R*np.arcsin(np.sqrt(a))

price_map = {"$":80, "$$":200, "$$$":400}
def pref_match_row(row):
    return int(pd.notnull(row["preferred_types"]) and str(row["category"]) in str(row["preferred_types"]))

# 3) ميزات
df["pref_match"] = df.apply(pref_match_row, axis=1)
df["approx_price"] = df["price_tier"].map(price_map).fillna(200)
df["budget_fit"]  = ((df["approx_price"]>=df["budget_min"]) & (df["approx_price"]<=df["budget_max"])).astype(int)
df["distance_km"] = haversine(df["home_latitude"], df["home_longitude"], df["latitude"], df["longitude"]).round(3)
df["hour"] = df["action_ts"].dt.hour
df["dow"]  = df["action_ts"].dt.weekday
df["is_weekend"] = (df["day_type"]=="Weekend").astype(int)
df["log_rating_count"] = np.log1p(df["rating_count"].fillna(0))
df["rating_avg_filled"] = df["rating_avg"].fillna(df["rating_avg"].median())
df["want_family"] = (df["marital_status"]=="Married").astype(int)
df["family_align"] = ((df["want_family"]==1) & (df["family_friendly"]==1)).astype(int)
df["outdoor_flag"] = df["outdoor"].fillna(0).astype(int)

for col in ["device","category","time_of_day"]:
    df[col] = df[col].astype("category").cat.codes

FEATURES = [
    "pref_match","budget_fit","distance_km","hour","dow","is_weekend",
    "rating_avg_filled","log_rating_count","family_align","outdoor_flag",
    "device","category","time_of_day","clicks","time_spent_seconds"
]

# 4) تقسيم زمني: آخر 20% اختبار
df_sorted = df.sort_values("action_ts")
cut = int(len(df_sorted)*0.8)
test_ids = set(df_sorted.iloc[cut:]["interaction_id"])
is_test = df["interaction_id"].isin(test_ids)

X_train, y_train = df.loc[~is_test, FEATURES].values, df.loc[~is_test, "label"].values
X_test,  y_test  = df.loc[ is_test, FEATURES].values, df.loc[ is_test, "label"].values

# 5) تدريب موديل أساس
clf = GradientBoostingClassifier(random_state=0)
clf.fit(X_train, y_train)
proba = clf.predict_proba(X_test)[:,1]

# 6) مقاييس + NDCG@10
auc = roc_auc_score(y_test, proba) if len(np.unique(y_test))>1 else float("nan")
ap  = average_precision_score(y_test, proba) if len(np.unique(y_test))>1 else float("nan")

def ndcg_at_k(labels, scores, k=10):
    if len(labels)==0: return np.nan
    order_idx = np.argsort(-scores)
    k_eff = min(k, len(order_idx))
    order = order_idx[:k_eff]
    denom = np.log2(np.arange(2, k_eff+2))
    gains = (2**labels[order] - 1)/denom
    ideal = np.sort(labels)[::-1][:k_eff]
    ideal_g = (2**ideal - 1)/denom
    return float(gains.sum() / (ideal_g.sum() if ideal_g.sum()>0 else 1.0))

tmp = df.loc[is_test, ["user_id","label"]].copy()
tmp["pred"] = proba
ndcgs = []
for uid, g in tmp.groupby("user_id"):
    labs, sc = g["label"].values, g["pred"].values
    if labs.sum()==0: continue
    ndcgs.append(ndcg_at_k(labs, sc, 10))
ndcg10 = float(np.nanmean(ndcgs)) if ndcgs else float("nan")

print({"AUC":round(auc,4), "AP":round(ap,4), "NDCG@10":round(ndcg10,4)})
```

```python
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix, roc_auc_score, average_precision_score

# درجات الاختبار موجودة من STEP 2
X_test = df.loc[is_test, FEATURES].values
y_test = df.loc[is_test, "label"].values
proba  = clf.predict_proba(X_test)[:,1]

# اختر عتبة (threshold) للتصنيف — 0.5 مبدئياً (عدّلها لاحقًا لتعظيم F1)
thr = 0.5
y_pred = (proba >= thr).astype(int)

# مقاييس تصنيف نقطية
prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average="binary", zero_division=0)

# مصفوفة الالتباس
cm = confusion_matrix(y_test, y_pred)  # [[TN, FP],[FN, TP]]

# مقاييس احتمالية (مستقلة عن العتبة)
auc = roc_auc_score(y_test, proba) if len(set(y_test))>1 else float('nan')
ap  = average_precision_score(y_test, proba) if len(set(y_test))>1 else float('nan')

print({
    "AUC": round(auc,4),
    "AveragePrecision(PR-AUC)": round(ap,4),
    "Precision@thr": round(prec,4),
    "Recall@thr": round(rec,4),
    "F1@thr": round(f1,4),
    "ConfusionMatrix": cm.tolist()
})
```

```python
# Threshold Tuning — نبحث عن أفضل عتبة تزيد F1 (وبنطبع المقاييس عندها)
import numpy as np
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix, roc_auc_score, average_precision_score

# نعيد حساب درجات الاختبار (لو كانت proba/y_test مو متاحة)
X_test = df.loc[is_test, FEATURES].values
y_test = df.loc[is_test, "label"].values
proba  = clf.predict_proba(X_test)[:,1]

def metrics_at_thr(y_true, y_score, thr):
    y_pred = (y_score >= thr).astype(int)
    prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    cm = confusion_matrix(y_true, y_pred)
    return prec, rec, f1, cm

# جرّب عتبات من 0.05 إلى 0.95 بخطوة 0.01
ths = np.round(np.arange(0.05, 0.951, 0.01), 2)
best = {"thr": None, "f1": -1, "prec": None, "rec": None, "cm": None}

for t in ths:
    prec, rec, f1, cm = metrics_at_thr(y_test, proba, t)
    if f1 > best["f1"]:
        best = {"thr": float(t), "f1": float(f1), "prec": float(prec), "rec": float(rec), "cm": cm.tolist()}

# مقاييس احتمالية (مستقلة عن العتبة) مرجعية
auc = roc_auc_score(y_test, proba) if len(set(y_test))>1 else float('nan')
ap  = average_precision_score(y_test, proba) if len(set(y_test))>1 else float('nan')

print({
    "BestThreshold": best["thr"],
    "F1@best": round(best["f1"],4),
    "Precision@best": round(best["prec"],4),
    "Recall@best": round(best["rec"],4),
    "ConfusionMatrix@best": best["cm"],
    "AUC": round(auc,4),
    "PR-AUC": round(ap,4)
})

# خزّن أحسن عتبة لاستخدامها لاحقًا
THR_BEST = best["thr"]
```

```python
# Hyperparameter Tuning for GradientBoostingClassifier (يعطيك موديل أدق بدون تعقيد)
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix, roc_auc_score, average_precision_score
import numpy as np

# بيانات التدريب/الاختبار من STEP 2
X_train = df.loc[~is_test, FEATURES].values
y_train = df.loc[~is_test, "label"].values
X_test  = df.loc[ is_test, FEATURES].values
y_test  = df.loc[ is_test, "label"].values

# شبكة بارامترات خفيفة (متوازنة بين الدقة والسرعة)
param_grid = {
    "n_estimators":   [100, 200],
    "learning_rate":  [0.05, 0.1, 0.2],
    "max_depth":      [2, 3],
    "subsample":      [0.8, 1.0],
    "min_samples_leaf":[1, 10],
}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
scoring = {
    "f1": "f1",
    "ap": "average_precision",
    "auc": "roc_auc"
}

gcv = GridSearchCV(
    estimator=GradientBoostingClassifier(random_state=0),
    param_grid=param_grid,
    scoring=scoring,
    refit="f1",        # نختار أفضل موديل حسب F1
    cv=cv,
    n_jobs=-1,
    verbose=0
)

gcv.fit(X_train, y_train)
best_clf = gcv.best_estimator_

# تقييم على الاختبار بالموديل المضبوط
proba_tuned = best_clf.predict_proba(X_test)[:, 1]
auc  = roc_auc_score(y_test, proba_tuned) if len(set(y_test))>1 else float('nan')
ap   = average_precision_score(y_test, proba_tuned) if len(set(y_test))>1 else float('nan')

# إعادة ضبط العتبة المثلى على الموديل الجديد
ths = np.round(np.arange(0.05, 0.951, 0.01), 2)
best = {"thr": None, "f1": -1, "prec": None, "rec": None, "cm": None}
for t in ths:
    y_pred = (proba_tuned >= t).astype(int)
    prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average="binary", zero_division=0)
    cm = confusion_matrix(y_test, y_pred)
    if f1 > best["f1"]:
        best = {"thr": float(t), "f1": float(f1), "prec": float(prec), "rec": float(rec), "cm": cm.tolist()}

print("Best Params:", gcv.best_params_)
print({
    "AUC": round(auc,4),
    "PR-AUC": round(ap,4),
    "BestThreshold": best["thr"],
    "F1@best": round(best["f1"],4),
    "Precision@best": round(best["prec"],4),
    "Recall@best": round(best["rec"],4),
    "ConfusionMatrix@best": best["cm"]
})

# استخدم الموديل والعتبة المضبوطة لاحقًا في التوصيات
clf = best_clf
THR_BEST = best["thr"]
```

```python
# STEP 3.5 — Recommend from ALL events (global candidate set)

import pandas as pd
import numpy as np

# 0) نشتق جدول الفعاليات من df (أو حمّل events.csv لو عندك)
event_cols = ["event_id","name","category","subcategory","city_event","location_name","address",
              "latitude","longitude","open_hours","rating_avg","rating_count","price_tier",
              "seasonality","family_friendly","outdoor","description"]
events_all = df[event_cols].drop_duplicates("event_id").reset_index(drop=True)

# 1) نحفظ ترميزات الفئات من بيانات التدريب (علشان نطابقها وقت inference)
cat_encoders = {}
for col in ["device","category","time_of_day"]:
    # ناخذ ترميز التدريب من df بعد step 2 (حصلنا على أرقام بالفعل هناك)
    # لذا نعيد بناء mapping من القيم النصية الأصلية -> الأكواد
    # نجيب زوج (القيمة النصية, الكود) من نسخة "قبل" الترميز:
    # نصنع قاموس من ترتيب الفئات كما كانت:
    series = df[col].copy()
    # لو كانت بالفعل أرقام (قد تكون codes) نتجاوز:
    if not np.issubdtype(series.dtype, np.number):
        cats = pd.Categorical(series).categories.tolist()
        cat_encoders[col] = {v:i for i,v in enumerate(cats)}
    else:
        # fallback: نحدد ماب افتراضي
        if col=="device":
            cat_encoders[col] = {"iOS":0,"Android":1,"Web":2}
        elif col=="time_of_day":
            cat_encoders[col] = {"Morning":0,"Afternoon":1,"Evening":2,"Night":3}
        elif col=="category":
            cat_encoders[col] = {v:i for i,v in enumerate(df["category"].astype("category").cat.categories.tolist())}

price_map = {"$":80, "$$":200, "$$$":400}

def haversine_np(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1; dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 2*R*np.arcsin(np.sqrt(a))

def _encode(col, value):
    enc = cat_encoders.get(col, {})
    return enc.get(value, -1)

def recommend_from_all_events(user_id, topN=10,
                              hour=None, day_type=None, device="Android",
                              max_distance_km=None, must_fit_budget=False,
                              prefer_family=None):
    """
    - hour: ساعة افتراضية للتنبؤ (0..23). لو None نأخذ ساعة متوسطة 18 (مساء).
    - day_type: "Weekday"/"Weekend"/"Holiday" (لـ is_weekend). لو None نستنتج من hour≈مساء يوم عمل.
    - device: "Android"/"iOS"/"Web".
    - max_distance_km: فلتر مسافة (مثلاً 40).
    - must_fit_budget: لو True لا يرشح إلا ما يطابق ميزانية المستخدم.
    - prefer_family: لو True يفضل العائلي، لو False يتجاهل، لو None يتركها للميزات فقط.
    """
    # نجيب بيانات المستخدم (أحدث صف له)
    u = df[df["user_id"]==user_id].iloc[0:1].copy()
    if u.empty:
        raise ValueError("User not found in df.")
    u = u.iloc[0]

    # سياق التنبؤ
    hour = 18 if hour is None else int(hour)
    time_of_day_val = ("Morning" if 5<=hour<12 else
                       "Afternoon" if 12<=hour<17 else
                       "Evening" if 17<=hour<23 else "Night")
    if day_type is None:
        day_type = "Weekday"
    is_weekend = int(day_type == "Weekend")

    # نبني ميزات لكل فعالية
    cand = events_all.copy()
    cand["pref_match"] = cand["category"].apply(lambda c: int(pd.notnull(u["preferred_types"]) and str(c) in str(u["preferred_types"])))
    cand["approx_price"] = cand["price_tier"].map(price_map).fillna(200)
    cand["budget_fit"] = ((cand["approx_price"]>=u["budget_min"]) & (cand["approx_price"]<=u["budget_max"])).astype(int)

    cand["distance_km"] = haversine_np(
        u["home_latitude"], u["home_longitude"],
        cand["latitude"], cand["longitude"]
    ).round(3)

    cand["hour"] = hour
    cand["dow"]  = 4  # تقدير بسيط (الخميس)؛ ممكن تمريره باراميتر لو حبيت
    cand["is_weekend"] = is_weekend

    cand["log_rating_count"] = np.log1p(cand["rating_count"].fillna(0))
    cand["rating_avg_filled"] = cand["rating_avg"].fillna(cand["rating_avg"].median())

    want_family = int(u["marital_status"]=="Married")
    cand["family_align"] = ((want_family==1) & (cand["family_friendly"]==1)).astype(int)
    cand["outdoor_flag"] = cand["outdoor"].fillna(0).astype(int)

    # ترميز الفئات ليتطابق مع التدريب
    cand["device"] = _encode("device", device)
    cand["time_of_day"] = _encode("time_of_day", time_of_day_val)
    cand["category_enc"] = cand["category"].apply(lambda x: _encode("category", x))

    # الأعمدة بنفس ترتيب FEATURES في التدريب (مع استبدال category بالترميز)
    FEATURES = [
        "pref_match","budget_fit","distance_km","hour","dow","is_weekend",
        "rating_avg_filled","log_rating_count","family_align","outdoor_flag",
        "device","category_enc","time_of_day"
    ]
    # ميزتي clicks و time_spent_seconds غير متاحة خارج التفاعلات؛ نضع قيمًا افتراضية معقولة
    cand["clicks"] = 1
    cand["time_spent_seconds"] = 60
    FEATURES += ["clicks","time_spent_seconds"]

    # فلاتر اختيارية
    if max_distance_km is not None:
        cand = cand[cand["distance_km"] <= max_distance_km]
    if must_fit_budget:
        cand = cand[cand["budget_fit"]==1]
    if prefer_family is True:
        cand = cand[cand["family_friendly"]==1]

    if cand.empty:
        return cand  # لا توجد مرشحات مناسبة

    # تنبؤ بالاحتمال باستخدام clf المدرّب في Step 2
    X = cand[FEATURES].values
    cand["score"] = clf.predict_proba(X)[:,1]

    cols_show = ["event_id","name","category","pref_match","budget_fit","distance_km",
                 "rating_avg_filled","family_friendly","outdoor","score"]
    return cand.sort_values("score", ascending=False).head(topN)[cols_show]

# ——— مثال استخدام ———
# غيّر user_id لأي مستخدم عندك:
uid = df["user_id"].iloc[0]
recs_all = recommend_from_all_events(
    user_id=uid,
    topN=10,
    hour=10,                # صباحاً
    day_type="Weekend",     # عطلة
    device="Android",
    max_distance_km=40,     # ضمن 40 كم
    must_fit_budget=True,   # يطابق الميزانية
    prefer_family=True      # يفضل العائلي
)
recs_all
```

```python
# تصحيح: مطابقة الميزانية باستخدام تداخل النطاقات بدل رقم ثابت
def ranges_overlap(a_min, a_max, b_min, b_max):
    return (max(a_min, b_min) <= min(a_max, b_max))

# نطاقات تقديرية مرنة (تقدر تعدلها لاحقاً)
tier_ranges = {
    "$":   (0,   150),
    "$$":  (100, 300),
    "$$$": (250, 600),
}

def compute_budget_fit_for_candidates(cand_df, user_min, user_max):
    tr_min = cand_df["price_tier"].map(lambda t: tier_ranges.get(t, (0,600))[0])
    tr_max = cand_df["price_tier"].map(lambda t: tier_ranges.get(t, (0,600))[1])
    # تداخل نطاق ميزانية المستخدم مع نطاق طبقة السعر
    return (np.minimum(tr_max, user_max) - np.maximum(tr_min, user_min) >= 0).astype(int)
```

```python
# خلية 2 — دوال مساعدة (لا تغيّر شيء هنا)
import numpy as np

# هافرساين لحساب المسافة بالكيلومترات
def haversine_np(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1; dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 2*R*np.arcsin(np.sqrt(a))

# نطاقات سعر مرنة لكل طبقة
tier_ranges = {
    "$":   (0,   150),
    "$$":  (100, 300),
    "$$$": (250, 600),
}

# مطابقة ميزانية المستخدم مع نطاق سعر الفعالية (تداخل نطاقات)
def compute_budget_fit_for_candidates(cand_df, user_min, user_max):
    tr_min = cand_df["price_tier"].map(lambda t: tier_ranges.get(t, (0,600))[0])
    tr_max = cand_df["price_tier"].map(lambda t: tier_ranges.get(t, (0,600))[1])
    return (np.minimum(tr_max, user_max) - np.maximum(tr_min, user_min) >= 0).astype(int)

# ترميزات بسيطة للسياق (تتوافق مع التدريب)
device_map = {"Android":0, "iOS":1, "Web":2}
tod_map    = {"Morning":0, "Afternoon":1, "Evening":2, "Night":3}
```

```python
def build_itinerary(user_id,
                    day_start,                 # pd.Timestamp("YYYY-MM-DD HH:MM")
                    max_spots=6,
                    include_breakfast=True,
                    include_lunch=True,
                    include_dinner=True,
                    max_distance_km=80,
                    must_fit_budget=True):
    """
    يُنتج جدول زمني مرتب + أزمنة انتقال + مدد الزيارة.
    """
    # 1) ترشيحات عامة من كل الفعاليات (كتالوج كامل)
    base_recs = recommend_from_all_events(
        user_id=user_id,
        topN=120,
        hour=day_start.hour,
        day_type="Weekend" if day_start.weekday() in (4,5) else "Weekday",
        device="Android",
        max_distance_km=max_distance_km,
        must_fit_budget=must_fit_budget,
        prefer_family=None
    )

    if base_recs is None or base_recs.empty:
        return pd.DataFrame(columns=["start_time","end_time","event_id","name","category",
                                     "distance_from_prev_km","travel_min","stay_min","score",
                                     "address","latitude","longitude"])

    # فلترة على العتبة المثلى (اختياريًا لتقليل الضجيج)
    recs = base_recs[base_recs["score"] >= THR_BEST].copy() if "THR_BEST" in globals() else base_recs.copy()
    if recs.empty:
        recs = base_recs.copy()

    # ألحق أعمدة الموقع/العنوان إن ما كانت موجودة
    if "latitude" not in recs.columns:
        recs = recs.merge(events_all[["event_id","address","latitude","longitude"]], on="event_id", how="left")

    # 2) فصل الطعام عن الأنشطة
    food_mask = recs["category"].str.contains("Food", case=False, na=False)
    foods = recs[food_mask].sort_values("score", ascending=False).copy()
    acts  = recs[~food_mask].sort_values("score", ascending=False).copy()

    # ✅ خلّي كل العناصر dicts (عشان ما نخلط Series مع dict)
    picks = []
    if include_breakfast and len(foods) >= 1: picks.append(foods.iloc[0].to_dict())
    if include_lunch     and len(foods) >= 2: picks.append(foods.iloc[1].to_dict())
    if include_dinner    and len(foods) >= 3: picks.append(foods.iloc[2].to_dict())

    need = max(0, max_spots - len(picks))
    if need > 0:
        picks += acts.head(need).to_dict(orient="records")

    cand = pd.DataFrame(picks).drop_duplicates("event_id").reset_index(drop=True)
    if cand.empty:
        cand = recs.head(max_spots).copy()

    # 3) ترتيب المسار بطريقة Nearest-Neighbor من موقع المستخدم
    u = df[df["user_id"]==user_id].iloc[0]
    cur_lat, cur_lon = u["home_latitude"], u["home_longitude"]
    ordered = []
    remaining = cand.copy()

    while not remaining.empty:
        dists = np.sqrt((remaining["latitude"]-cur_lat)**2 + (remaining["longitude"]-cur_lon)**2)
        idx = dists.idxmin()
        chosen = remaining.loc[idx]
        ordered.append(chosen.to_dict())   # نحافظ على dict
        cur_lat, cur_lon = chosen["latitude"], chosen["longitude"]
        remaining = remaining.drop(index=idx)

    plan = pd.DataFrame(ordered).reset_index(drop=True)

    # 4) حساب أوقات البدء/الانتهاء + السفر + مدة الزيارة
    cur_lat, cur_lon = u["home_latitude"], u["home_longitude"]
    t = pd.to_datetime(day_start).replace(second=0, microsecond=0)

    start_times, end_times, travel_mins, stay_mins, dist_prev = [], [], [], [], []

    def _travel_time_min(lat1, lon1, lat2, lon2, speed_kmh=30):
        R = 6371.0
        φ1, λ1, φ2, λ2 = map(np.radians, [lat1, lon1, lat2, lon2])
        dφ = φ2-φ1; dλ = λ2-λ1
        a = np.sin(dφ/2)**2 + np.cos(φ1)*np.cos(φ2)*np.sin(dλ/2)**2
        dist_km = 2*R*np.arcsin(np.sqrt(a))
        return max(5, (dist_km / max(5, speed_kmh)) * 60)

    for _, r in plan.iterrows():
        tr_min = _travel_time_min(cur_lat, cur_lon, r["latitude"], r["longitude"])
        dist_km = max(0.0, float(tr_min) * 30.0 / 60.0)  # تقريب المسافة من الزمن/السرعة
        t = t + timedelta(minutes=tr_min)
        start_times.append(t)

        stay = 60 if str(r["category"]).lower().startswith("food") else 90
        end = t + timedelta(minutes=stay)
        end_times.append(end)
        stay_mins.append(stay)
        travel_mins.append(int(round(tr_min)))
        dist_prev.append(round(dist_km, 2))

        t = end
        cur_lat, cur_lon = r["latitude"], r["longitude"]

    plan["start_time"] = start_times
    plan["end_time"]   = end_times
    plan["travel_min"] = travel_mins
    plan["stay_min"]   = stay_mins
    plan["distance_from_prev_km"] = dist_prev

    cols = ["start_time","end_time","event_id","name","category",
            "distance_from_prev_km","travel_min","stay_min","score","address","latitude","longitude"]
    return plan[cols]
```

```python
# تصحيح parsing لساعات العمل: نحول "10:00-22:00" -> ("10:00","22:00") ثم نعيد بناء الخريطة ونطبق
import ast, pandas as pd, numpy as np

def _parse_hours(hs):
    default = {"Daily": ("00:00","23:59")}
    if pd.isna(hs):
        return default
    try:
        d = ast.literal_eval(str(hs))
        if not isinstance(d, dict):
            return default
        out = {}
        for k, v in d.items():
            if isinstance(v, str):
                s = v.replace(" ", "")
                parts = s.split("-")
                if len(parts) == 2:
                    out[k] = (parts[0], parts[1])
                else:
                    out[k] = ("00:00", "23:59")
            elif isinstance(v, (list, tuple)) and len(v) == 2:
                out[k] = (str(v[0]), str(v[1]))
            else:
                out[k] = ("00:00", "23:59")
        return out or default
    except Exception:
        return default

# إعادة بناء خريطة ساعات العمل
_hours_map = {}
for _, r in events_all[["event_id","open_hours"]].drop_duplicates("event_id").iterrows():
    _hours_map[int(r["event_id"])] = _parse_hours(r["open_hours"])

# تطبيقها مجددًا
itinerary_hours = apply_open_hours(uid, day_start, itinerary)
itinerary_hours
```

```python
# === Patch: 2-opt route optimization (robust) + re-apply open hours ===
import numpy as np
import pandas as pd
from itertools import combinations

# تأكد أن الإحداثيات والعنوان موجودة في itinerary_hours
needed_cols = {"event_id","address","latitude","longitude"}
if not needed_cols.issubset(set(itinerary_hours.columns)):
    itinerary_hours = itinerary_hours.merge(
        events_all[["event_id","address","latitude","longitude"]],
        on="event_id", how="left"
    )

def _travel_time_min(lat1, lon1, lat2, lon2, speed_kmh=30):
    R = 6371.0
    φ1, λ1, φ2, λ2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dφ = φ2-φ1; dλ = λ2-λ1
    a = np.sin(dφ/2)**2 + np.cos(φ1)*np.cos(φ2)*np.sin(dλ/2)**2
    dist_km = 2*R*np.arcsin(np.sqrt(a))
    return max(5, (dist_km / max(5, speed_kmh)) * 60), dist_km

def _pairwise_matrix(points, start_lat, start_lon):
    n = len(points)
    time_mat = np.zeros((n+1, n+1))
    dist_mat = np.zeros((n+1, n+1))
    for i, p in enumerate(points):
        t, d = _travel_time_min(start_lat, start_lon, p["latitude"], p["longitude"])
        time_mat[-1, i] = t; dist_mat[-1, i] = d
        time_mat[i, -1] = t; dist_mat[i, -1] = d
    for i in range(n):
        for j in range(n):
            if i == j: continue
            t, d = _travel_time_min(points[i]["latitude"], points[i]["longitude"],
                                    points[j]["latitude"], points[j]["longitude"])
            time_mat[i, j] = t; dist_mat[i, j] = d
    return time_mat, dist_mat

def _route_cost(order, time_mat):
    cost = time_mat[-1, order[0]]
    for i in range(len(order)-1):
        cost += time_mat[order[i], order[i+1]]
    return cost

def two_opt(order, time_mat, max_iter=200):
    best = order[:]
    best_cost = _route_cost(best, time_mat)
    improved = True
    it = 0
    while improved and it < max_iter:
        improved = False
        it += 1
        for i, j in combinations(range(1, len(order)-1), 2):
            if j <= i:
                continue
            new = best[:i] + best[i:j][::-1] + best[j:]
            new_cost = _route_cost(new, time_mat)
            if new_cost + 1e-6 < best_cost:
                best, best_cost = new, new_cost
                improved = True
        if not improved:
            break
    return best, best_cost

def optimize_itinerary_order_2opt(itinerary_df, user_id):
    u = df[df["user_id"]==user_id].iloc[0]
    start_lat, start_lon = float(u["home_latitude"]), float(u["home_longitude"])
    base_cols = ["event_id","name","category","address","latitude","longitude","score"]
    pts_df = itinerary_df[base_cols].reset_index(drop=True).copy()
    pts = pts_df.to_dict(orient="records")
    T, D = _pairwise_matrix(pts, start_lat, start_lon)
    init_order = list(range(len(pts)))
    best_order, best_cost = two_opt(init_order, T)
    opt_pts = [pts[i] for i in best_order]
    return pd.DataFrame(opt_pts), best_order, best_cost

# شغّل التحسين
opt_points_df, best_order, best_travel_cost = optimize_itinerary_order_2opt(itinerary_hours, uid)

# جهّز جدول بسيط مطابق لتوقّعات apply_open_hours (لا يحتاج start/end ولا أعمدة المسافة/الزمن)
opt_itinerary_like = opt_points_df[["event_id","name","category","address","latitude","longitude","score"]].copy()

# أعد تطبيق ساعات العمل على الترتيب المحسّن
itinerary_optimized = apply_open_hours(uid, day_start, opt_itinerary_like)

summary_before = (int(itinerary_hours["travel_min"].sum()), int(itinerary_hours["stay_min"].sum()))
summary_after  = (int(itinerary_optimized["travel_min"].sum()), int(itinerary_optimized["stay_min"].sum()))
print({
    "before": {"travel_min": summary_before[0], "stay_min": summary_before[1]},
    "after":  {"travel_min": summary_after[0],  "stay_min": summary_after[1]}
})
itinerary_optimized
```

```python
# === build_itinerary (نسخة تستخدم MLP عبر recommend_from_all_events_ai) ===
import pandas as pd
import numpy as np
from datetime import timedelta

# حد أدنى بسيط لحساب زمن الانتقال بالدقائق
def _travel_time_min(lat1, lon1, lat2, lon2, speed_kmh=30):
    R = 6371.0
    φ1, λ1, φ2, λ2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dφ = φ2-φ1; dλ = λ2-λ1
    a = np.sin(dφ/2)**2 + np.cos(φ1)*np.cos(φ2)*np.sin(dλ/2)**2
    dist_km = 2*R*np.arcsin(np.sqrt(a))
    return max(5, (dist_km / max(5, speed_kmh)) * 60)  # حد أدنى 5 دقائق

def build_itinerary(user_id,
                    day_start,                 # pd.Timestamp("YYYY-MM-DD HH:MM")
                    max_spots=6,
                    include_breakfast=True,
                    include_lunch=True,
                    include_dinner=True,
                    max_distance_km=80,
                    must_fit_budget=True):
    """
    يُنتج جدول زمني مرتب + أزمنة انتقال + مدد الزيارة باستخدام MLP (ReLU).
    يتطلب أن تكون recommend_from_all_events_ai/events_all/df معرفة مسبقاً.
    """
    # 1) ترشيحات عامة من كامل الكتالوج باستخدام MLP
    base_recs = recommend_from_all_events_ai(
        user_id=user_id,
        topN=120,
        hour=day_start.hour,
        day_type="Weekend" if day_start.weekday() in (4,5) else "Weekday",
        device="Android",
        max_distance_km=max_distance_km,
        must_fit_budget=must_fit_budget,
        prefer_family=None,
        model_type="mlp",
        use_threshold=True
    )

    if base_recs is None or base_recs.empty:
        return pd.DataFrame(columns=["start_time","end_time","event_id","name","category",
                                     "distance_from_prev_km","travel_min","stay_min","score",
                                     "address","latitude","longitude"])

    # ألحق الإحداثيات/العنوان إن لزم
    if "latitude" not in base_recs.columns:
        base_recs = base_recs.merge(
            events_all[["event_id","address","latitude","longitude"]],
            on="event_id", how="left"
        )

    # 2) فصل الطعام عن الأنشطة
    food_mask = base_recs["category"].str.contains("Food", case=False, na=False)
    foods = base_recs[food_mask].sort_values("score", ascending=False).copy()
    acts  = base_recs[~food_mask].sort_values("score", ascending=False).copy()

    picks = []
    if include_breakfast and len(foods) >= 1: picks.append(foods.iloc[0].to_dict())
    if include_lunch     and len(foods) >= 2: picks.append(foods.iloc[1].to_dict())
    if include_dinner    and len(foods) >= 3: picks.append(foods.iloc[2].to_dict())
    need = max(0, max_spots - len(picks))
    if need > 0:
        picks += acts.head(need).to_dict(orient="records")

    cand = pd.DataFrame(picks).drop_duplicates("event_id").reset_index(drop=True)
    if cand.empty:
        cand = base_recs.head(max_spots).copy()

    # 3) ترتيب المسار بـ Nearest-Neighbor من موقع المستخدم
    u = df[df["user_id"]==user_id].iloc[0]
    cur_lat, cur_lon = float(u["home_latitude"]), float(u["home_longitude"])
    ordered = []
    remaining = cand.copy()

    while not remaining.empty:
        dists = np.sqrt((remaining["latitude"]-cur_lat)**2 + (remaining["longitude"]-cur_lon)**2)
        idx = dists.idxmin()
        chosen = remaining.loc[idx]
        ordered.append(chosen.to_dict())
        cur_lat, cur_lon = float(chosen["latitude"]), float(chosen["longitude"])
        remaining = remaining.drop(index=idx)

    plan = pd.DataFrame(ordered).reset_index(drop=True)

    # 4) حساب أوقات البدء/الانتهاء + السفر + مدة الزيارة
    cur_lat, cur_lon = float(u["home_latitude"]), float(u["home_longitude"])
    t = pd.to_datetime(day_start).replace(second=0, microsecond=0)

    start_times, end_times, travel_mins, stay_mins, dist_prev = [], [], [], [], []

    for _, r in plan.iterrows():
        tr_min = _travel_time_min(cur_lat, cur_lon, float(r["latitude"]), float(r["longitude"]))
        dist_km = max(0.0, float(tr_min) * 30.0 / 60.0)  # تقريب المسافة من الزمن/السرعة
        t = t + timedelta(minutes=tr_min)
        start_times.append(t)

        stay = 60 if str(r["category"]).lower().startswith("food") else 90
        end = t + timedelta(minutes=stay)
        end_times.append(end)
        stay_mins.append(stay)
        travel_mins.append(int(round(tr_min)))
        dist_prev.append(round(dist_km, 2))

        t = end
        cur_lat, cur_lon = float(r["latitude"]), float(r["longitude"])

    plan["start_time"] = start_times
    plan["end_time"]   = end_times
    plan["travel_min"] = travel_mins
    plan["stay_min"]   = stay_mins
    plan["distance_from_prev_km"] = dist_prev

    cols = ["start_time","end_time","event_id","name","category",
            "distance_from_prev_km","travel_min","stay_min","score","address","latitude","longitude"]
    return plan[cols]
```
